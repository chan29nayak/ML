```python
X = df.x

y = df.y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size =0.7, random_state=175)
```

```python
X_train = X_train[:, np.newaxis]
X_test = X_test[:, np.newaxis]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
c = [i for i in range(1,91,1)]
fig = plt.figure()
plt.plot(c,y_test, color="red", linewidth=2.5, linestyle="-")
plt.plot(c,y_pred, color="blue",  linewidth=2.5, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)   
plt.xlabel('x', fontsize=18)                       
plt.ylabel('y', fontsize=16)                       
```

```python
c = [i for i in range(1,91,1)]
fig = plt.figure()
plt.plot(c,y_test-y_pred, color="red", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)               
plt.xlabel('x', fontsize=18)                      
plt.ylabel('ytest-ypred', fontsize=16)                
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
```

```python
r_squared = r2_score(y_test, y_pred)
```

```python
print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```
